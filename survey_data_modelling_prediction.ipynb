```python
#label encoding the age_group column
from sklearn.preprocessing import LabelEncoder
labelenocoder=LabelEncoder()
df['age_group']=labelenocoder.fit_transform(df['age_group'])
df['income_levels']=labelenocoder.fit_transform(df['income_levels'])
df['health_concerns']=labelenocoder.fit_transform(df['health_concerns'])
df['consume_frequency(weekly)']=labelenocoder.fit_transform(df['consume_frequency(weekly)'])
df['preferable_consumption_size']=labelenocoder.fit_transform(df['preferable_consumption_size'])
```

```python
# List of all columns
all_columns = [
    'respondent_id', 'gender', 'zone', 'occupation', 'income_levels', 
    'consume_frequency(weekly)', 'current_brand', 'preferable_consumption_size', 
    'awareness_of_other_brands', 'reasons_for_choosing_brands', 'flavor_preference', 
    'purchase_channel', 'packaging_preference', 'health_concerns', 
    'typical_consumption_situations', 'age_group', 
    'awareness_of_other_brands_mapped', 'cf_ab_score', 'zas_score', 'bsi'
]

# Define columns to one-hot encode
columns_to_one_hot_encode = [
    col for col in all_columns if col not in [
        'age_group_encoded', 'income_levels_encoded', 'health_concerns_encoded', 
        'consume_frequency(weekly)_encoded', 'price_range_encoded', 'cf_ab_score', 
        'zas_score', 'bsi', 'respondent_id'
    ]
]

# Apply one-hot encoding
df = pd.get_dummies(df, columns=columns_to_one_hot_encode, drop_first=True)

# Display the first few rows of the updated DataFrame
df.head()
```

```python
target_column = 'price_range'  # Replace with actual target column
df['price_range'] = labelenocoder.fit_transform(df[target_column])
```

```python
# Define the columns to be used as features
X = df.drop(columns=['respondent_id', 'price_range'], axis=1)

# Define the target variable
Y = df['price_range']

# Verify the shapes of X and Y
print("Shape of Features (X):", X.shape)
print("Shape of Target (Y):", Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

GuassianNB For Classification

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Initialize and train the Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=labelenocoder.classes_))
```

Logisitic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred))
```

Support Vector Machines

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Initialize and train the SVM model (with 'rbf' kernel by default)
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

# Predict on the test set
y_pred_svm = svm_model.predict(X_test)

# Evaluate the SVM model
print("SVM Accuracy Score:", accuracy_score(y_test, y_pred_svm))
print("\nSVM Classification Report:\n", classification_report(y_test, y_pred_svm, target_names=labelenocoder.classes_))
```

Random Forest Classifier

```python
# Initialize and train the RandomForest model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=labelenocoder.classes_))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

models = [
    (
        "Logistic Regression",
        {"C": 1, "solver": 'liblinear'},
        LogisticRegression(C=1, solver='liblinear'),
        (X_train, y_train),
        (X_test, y_test),
    ),
    (
        "Random Forest",
        {"n_estimators": 30, "max_depth": 3},
        RandomForestClassifier(n_estimators=30, max_depth=3, random_state=42),
        (X_train, y_train),
        (X_test, y_test),
    ),
    (
        "SVM",
        {"C": 1, "kernel": 'linear'},
        SVC(C=1, kernel='linear', random_state=42),
        (X_train, y_train),  # SVM often requires scaled data
        (X_test, y_test),
    ),
    (
        "XGBClassifier",
        {"use_label_encoder": False, "eval_metric": 'logloss'},
        XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
        (X_train, y_train),
        (X_test, y_test),
    ),
    (
        "LightGBM Classifier",
        {"boosting_type": 'gbdt', "n_estimators": 50},
        LGBMClassifier(boosting_type='gbdt', n_estimators=50, random_state=42),
        (X_train, y_train),
        (X_test, y_test),
    ),
]
```
